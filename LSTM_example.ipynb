# LSTM with simple example:
* Long Short Term Memory networks – usually just called “LSTMs” – are a special kind of RNN, capable of learning long-term dependencies.
* It can be used for sequence prediction
* In this project we will try to train our LSTM model to predict the following sequence:
    ### 4,5,6,7,8,? 
    ### output = 9


```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

## Data Preparation
* make 100 data with 5 sequences 
* And target is the array of the output(6th sequence)
* Divide data by 100 for normalization

```python
Data = [[[(i+j)/100] for j in range(5)] for i in range(100)]
Target = [(i+5)/100 for i in range(100)]
```

```python
Data = np.array(Data, dtype = float)
Target = np.array(Target)
```

```python
Data.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(Data,Target,test_size = 0.2, random_state = 4)
```

## Creating Model

```python
model = Sequential()
model.add(LSTM((1),batch_input_shape=(None,5,1),return_sequences = False))
```

```python
model.compile(loss = 'mean_absolute_error', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train,y_train, epochs = 500, validation_data = (x_test,y_test))
```

```python
result = model.predict(x_test)
```

```python
plt.scatter(range(20),result,c='r')
plt.scatter(range(20),y_test, c = 'g')
plt.show()
```

```python
plt.plot(history.history['loss'])
```
